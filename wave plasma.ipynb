```python
import numpy
import math
from matplotlib import pyplot
%matplotlib inline
```

```python
te = 3*11600
ti = 1*11600
no = 10**17
K_b = 1.38*10**-23
e_o = 8.854*10**-12
e = 1.602*10**-19
m = 1.7*10**-27
me = 9.1*10**-31
gamma = 3.
```

```python
ld = (e_o*K_b*te)/(e**2*no)
ld
```

```python
L = 10**6
i = numpy.linspace(0.0,L,100)
```

```python
omega = i*numpy.sqrt(((K_b*te)/m)*(1/(1+i**2*ld)))


#omega_p = (no*e**2*i**2)/(e_o*me*i**2)
#print(i,omega)
```

```python
omega_p = numpy.sqrt((no*e**2*i**2)/(e_o*m*i**2))
```

```python
pyplot.figure(figsize=(20,10))
pyplot.grid(True)
pyplot.plot(i,omega,color='blue', ls ='--',linewidth='2', marker ='o')
pyplot.plot(i,omega_p,color='red',linewidth='2')
pyplot.xlabel('K ($m^{-1}$)', fontsize=20)
pyplot.ylabel('$\omega (Hz)$', fontsize=20)
pyplot.text(0,4.12922257e+08,'$\Omega_p$', fontsize=20)
#plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
pyplot.title('Dispersion relation for ion plasma',fontsize=20)


#pyplot.axhline(i=414141.41414141, omega=4.12797700e+08, linewidth=2, color='k')
#pyplot.xlim(0,100)
```

### Plasma wave for electron!

```python
c = (no*e**2)/(me*e_o)
c

d = (3*K_b*te)/me

print(c)
print(d)
```

```python
f = 100
f_i = numpy.linspace(0.0,f,100)


omega_i = numpy.sqrt(c + d*(f_i)**2)
#omega_i
```

```python
pyplot.figure(figsize=(20,10))
pyplot.grid(True)
pyplot.plot(i,omega_i,color='green', ls ='--',linewidth='2', marker ='o')
pyplot.xlabel('K ($m^{-1}$)', fontsize=20)
pyplot.ylabel('$\omega (Hz)$', fontsize=20)

pyplot.title('Dispersion relation for electron plasma',fontsize=20)
```

### Upper Hybrid Frequencies

```python
B = 0.1
b = numpy.linspace(0.001,B,25)
omega_h = numpy.sqrt(((no*e**2)/(me*e_o)) + ((e*b)/(me))**2)
#print(b)
```

```python
pyplot.figure(figsize=(15,6))
pyplot.grid(True)
pyplot.xlim(0,0.1)
pyplot.plot(b,omega_h,color='blue', ls ='--',linewidth='2', marker ='o')
pyplot.xlabel('B ($T$)', fontsize=20)
pyplot.ylabel('$\omega_h (Hz)$', fontsize=20)

pyplot.title('Upper Hybrid Frequency w.r.t Magnetic Field',fontsize=20)
```

### Lower Hybrid Frequency

```python
omega_l = numpy.sqrt(((e*b)/(me))*numpy.sqrt((no*e**2)/(e_o*m)))
pyplot.figure(figsize=(15,6))
pyplot.grid(True)
pyplot.xlim(0,0.1)
pyplot.plot(b,omega_l,color='red', ls ='--',linewidth='2', marker ='o')
pyplot.xlabel('B ($T$)', fontsize=20)
pyplot.ylabel('$\omega_l (Hz)$', fontsize=20)

pyplot.title('lower Hybrid Frequency w.r.t Magnetic Field',fontsize=20)
```
