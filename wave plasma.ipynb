```python
import numpy
import math
from matplotlib import pyplot
%matplotlib inline
```

```python
te = 3*11600
ti = 1*11600
no = 10**17
K_b = 1.38*10**-23
e_o = 8.854*10**-12
e = 1.602*10**-19
m = 1.7*10**-27
me = 9.1*10**-31
gamma = 3.
```

```python
ld = (e_o*K_b*te)/(e**2*no)
ld
```

```python
L = 10**6
i = numpy.linspace(0.0,L,100)
```

```python
omega = i*numpy.sqrt(((K_b*te)/m)*(1/(1+i**2*ld)))
omega

print(i,omega)
```

```python
pyplot.figure(figsize=(20,10))
pyplot.grid(True)
pyplot.plot(i,omega,color='blue', ls ='--',linewidth='2', marker ='o')
#pyplot.axhline(i=414141.41414141, omega=4.12797700e+08, linewidth=2, color='k')
#pyplot.xlim(0,100)
```
