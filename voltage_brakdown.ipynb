```python
import numpy
import math
from matplotlib import pyplot as plt
%matplotlib inline
```

The Breakdown voltage is given by the equation:

$V_B$ = $\frac{Bpd}{\ln(Apd)-\ln[\ln(1+\frac{1}{\gamma})]}$

### Helium

```python
A = 2.8
B = 34
gamma = 0.12
pd = 3
N = 50
n = numpy.linspace(0.8, pd, N)
```

```python
Vh= (B*n)/(numpy.log(A*n)-numpy.log(numpy.log(1+(1/gamma))))
```

```python
Aar = 13.6
Bar = 235
gamma = 0.12
pd = 3
N = 50
n = numpy.linspace(0.2, pd, N)
```

```python
Var= (Bar*n)/(numpy.log(Aar*n)-numpy.log(numpy.log(1+(1/gamma))))
```

```python
Aai = 14.6
Bai = 365
gamma = 0.12
pd = 3
N = 50
n = numpy.linspace(0.2, pd, N)
```

```python
Vai= (Bai*n)/(numpy.log(Aai*n)-numpy.log(numpy.log(1+(1/gamma))))
```

```python

plt.figure(figsize=(10,5))
plt.grid(True)
plt.xlabel('pd $(torr.cm)$', fontsize=16)
plt.ylabel('Breakdown voltage $V$', fontsize=16)
plt.xlim(0.,3.)
plt.ylim(-0.5,1000.)
plt.plot(n, Vh, color='blue', ls ='--',linewidth='2', marker ='o',label='Helium')
plt.plot(n, Vai, color='green', ls ='--',linewidth='2', marker ='o',label='air')
plt.plot(n, Var, color='red', ls ='--',linewidth='2', marker ='o',label='argon')
plt.title('Breakdown voltage $vs$ pd for He, Ar, Air',fontsize=20)

plt.legend(loc='upper right')
#plt.plot (n, V)
```
